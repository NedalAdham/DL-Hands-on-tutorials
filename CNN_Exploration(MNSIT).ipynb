```python
import numpy as np
from random import randint
import pandas as pd
import random 
import time
from keras.layers import Dropout
from sklearn.utils import shuffle
from collections import namedtuple, OrderedDict
from sklearn.preprocessing import MinMaxScaler
from tensorflow import keras
from IPython.display import display, clear_output
from tensorflow.keras.models import Sequential
import tensorflow as tf
from tensorflow.keras import optimizers
from sklearn.metrics import  confusion_matrix
from tensorflow import keras
from itertools import product
import tensorflow as tf
from tensorflow .keras .preprocessing import image
from tensorflow .keras .preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential , Model
from tensorflow.keras.layers import Dense,Activation,Flatten,BatchNormalization,Conv2D,MaxPool2D
from tensorflow.keras.optimizers import Adam
from tensorflow . keras . metrics import categorical_crossentropy 
import itertools
import matplotlib.pyplot as plt
import shutil 
import glob
import warnings
warnings.simplefilter(action='ignore' , category=FutureWarning)
```

```python
from keras.datasets import mnist 
from tensorflow.keras.utils import to_categorical
input_shape = (28, 28, 1)
(xtrain , y_train) , (xtest,y_test) = mnist.load_data()
print("xtrain shape", xtrain.shape)
print("y_train shape", y_train.shape)
print("X_test shape", xtest.shape)
print("y_test shape", y_test.shape)

# Scale images to the [0, 1] range
xtrain = xtrain.astype("float32") / 255
xtest = xtest.astype("float32") / 255
# Make sure images have shape (28, 28, 1)
xtrain = np.expand_dims(xtrain, -1)
xtest = np.expand_dims(xtest, -1)
print("xtrain shape:", xtrain.shape)
print(xtrain.shape[0], "train samples")
print(xtest.shape[0], "test samples")


# convert class vectors to binary class matrices
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
print ("")
print("xtrain shape", xtrain.shape)
print("y_train shape", y_train.shape)
print("X_test shape", xtest.shape)
print("y_test shape", y_test.shape)
```

```python
class RunBuilder ():
    @staticmethod
    def get_runs(params):
        Run = namedtuple('Run' , params.keys())
        
        runs = []
        
        for v in product (*params.values()):
            runs.append(Run(*v))
            
        return  runs
```

# Testing Batch number and Epoches

### Here we start with the simple model on Keras web site , now we will choose the best Number of epoches with constant batch size equal 32 and 64*

```python
epoch = np.arange(10,16)

batch_size = [32 , 64]
params = dict (epoch = epoch  , batch_size =batch_size)
runs = RunBuilder.get_runs(params)
display( epoch, len(runs) , runs)
```

```python
for i in range (len(runs)):

    run = runs[i]
    model =Sequential([
            keras.Input(input_shape),
            Conv2D(32, kernel_size=(3, 3), activation="relu"),
            MaxPool2D(pool_size=(2, 2)),
            Conv2D(64, kernel_size=(3, 3), activation="relu"),
            MaxPool2D(pool_size=(2, 2)),
            Flatten(),
            Dense(10, activation="softmax")])

    print("")
    print (f"""Model Number {i}
              with batch size = {run.batch_size}
              & number of epochs = {run.epoch}""")
    

    start = time.clock() 
    model.compile(loss="categorical_crossentropy", optimizer="SGD", metrics=["accuracy"])
    history = model.fit(xtrain, y_train, batch_size=run.batch_size, epochs=run.epoch, validation_split=0.1)
    end = time.clock()

    if i == 0:
        df = pd.DataFrame(columns = ['run', "batch_size", "number_of_epoch" , 'accuracy' ,"Val _accuracy" ,"time"])
    run_params = []
    results = OrderedDict()

    
    results["run"] = i
    results["batch_size"] = run.batch_size
    results['number_of_epoch'] = run.epoch
    results['accuracy'] = history.history['accuracy'][-1] * 100
    results["Val _accuracy"] = history.history['val_accuracy'][-1]*100
    results["time"] = end
    df = df .append([results] ,ignore_index = True)

    
    
```
